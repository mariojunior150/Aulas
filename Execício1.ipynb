### Calculadora de IMC

```python
peso = float(input('Digite seu peso (kg): '))
altura = float(input('Digite sua altura (m): '))

imc = peso / (altura ** 2)
print (f'Seu IMC é: {imc:.2f}')
print (f'Seu peso ideal é: {22 * (altura ** 2)}')
```

### Lógica booleana

```python
temperatura = float(input('Digite sua temperatura: '))
if temperatura >= 35.9 and temperatura <= 37.5:
  print ('Sua temperatura está normal')
elif temperatura < 35.9 or temperatura > 41.5:
  print ('Temperatura fora das condições aceitáveis, tente novamente.')
else:
  vacinado = input('Você está vacinado? (s/n): ')
  if vacinado not in ('s', 'n'):
    print ('Você digitou algo diferente de (s/n) tente novamente')
  elif vacinado == 'n':
    print(f'Você precisa procurar um médico pois está com febre ({temperatura}°) e não foi vacinado')
  else:
    print('Fique atento com a sua febre mesmo já estando vacinado')
```
